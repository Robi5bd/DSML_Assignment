### OneHotEncoder

```python
import pandas as pd
data=pd.read_csv('insurance.csv')
data
```

```python
from sklearn.preprocessing import OneHotEncoder    #import library
```

```python
# Create a copy of the dataset to avoid modifying the original data
data_onehot = data.copy()
```

```python
# Create a OneHotEncoder object
onehot_encoder = OneHotEncoder()
onehot_encoder
```

```python
# Apply one-hot encoding to 'sex', 'smoker', and 'region' columns
encoded_columns = pd.DataFrame(onehot_encoder.fit_transform(data[['sex', 'smoker', 'region']]).toarray(), 
                               columns=['female', 'male', 'non-smoker', 'smoker', 'northeast', 'northwest', 'southeast', 'southwest'])
```

```python
encoded_columns
```

```python
# Concatenate the one-hot encoded columns with the original dataset
data_onehot = pd.concat([data_onehot, encoded_columns], axis=1)
```

```python
data_onehot
```

```python
# Drop the original categorical columns
data_onehot.drop(['sex', 'smoker', 'region'], axis=1, inplace=True)
```

```python
# Display the updated dataset with one-hot encoding
data_onehot
```
